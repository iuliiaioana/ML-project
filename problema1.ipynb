```python
import pandas as pd
import re
import wordninja
from nltk.stem.wordnet import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# aici importez tot ce am nevoie
```

```python
def get_data(filename):
    train_tweets = pd.read_csv(filename)

    labels = train_tweets['label'].values
    print(labels)
    tweets = train_tweets['tweet'].values
    print(tweets)

    def prelucare(tweet):
        
        tweet = " ".join(filter(lambda x: x[0] != '@', tweet.split()))
        # aici scot toti @user
        result = re.sub(r"\W", "", tweet, flags=re.I)
        #aici scot tot ce nu e alphanumeric
       
        l=wordninja.split(result)
        # aici avem o lista cu toate cuvintele dupa spargerea cu wordninja
        
        stop_words = set(stopwords.words('english'))
        # aici avem toate stopwordurile din engl
        filtered_sentence = [w for w in l if not w in stop_words]
        # aici avem lista cu propozitia fara stop word,si fara caractere nonalphanumerice si fara "@user"
        
        
        return filtered_sentence

    def normalization(tweet_list):
        lem = WordNetLemmatizer()
        normalized_tweet = []
        for word in tweet_list:
            normalized_text = lem.lemmatize(word, 'v')
            normalized_tweet.append(normalized_text)
        return normalized_tweet
#     cu aceasta functie fac normalizarea textului

    shuffle_stratified = StratifiedShuffleSplit(n_splits=1, test_size=0.2)
#     impartim datele

    for train_index, test_index in shuffle_stratified.split(tweets, labels):
        
        msg_train, msg_test = tweets[train_index], tweets[test_index]

        labels_train, labels_test = labels[train_index], labels[test_index]

        
        
    print('de aici incep datele de train')
    
    for id in range(len(msg_train)):
        msg_train[id] =' '.join(normalization(prelucare(msg_train[id])))
        print(msg_train[id])
        
        
        
    print('de aici incep datele de test:')
    
    for id in range(len(msg_test)):
        msg_test[id] = ' '.join(normalization(prelucare(msg_test[id])))
        print(msg_train[id])

    return msg_train, labels_train, msg_test, labels_test
```

```python
# urmeaza sa apelez functia
# printurile dininteriorul forului sunt pentru a arata cum arata datele nu este necesar sa printam
```

```python
msg_train, y_train, msg_test, y_test = get_data('train1.csv')
```

```python
count_vectorizer = CountVectorizer(lowercase=True, analyzer='word', stop_words='english')
#vreau sa ne numere cuvintele , stopwords=elimina cuv de legatura din stringurile nst

count_vectorizer.fit(msg_train)
X_train = count_vectorizer.transform(msg_train)
#metoda de antrenare -- matricea de featururi

X_test = count_vectorizer.transform(msg_test)

model = MultinomialNB(alpha=0.01)
print('acum afisam x_train')
print(X_train)

model.fit(X_train,y_train)
print('acum printez y_train')
print(y_train)
#y este labalul
#x matricea sparsa


predictions = model.predict(X_test)


print(accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
print(model.predict_proba(X_test[0]))
print('matricea de confuzie')
print(confusion_matrix(y_test,predictions))
```
